```python
df['start_time_ts']=pd.to_datetime(df['start_time_ts'])
df['end_time_ts']=pd.to_datetime(df['end_time_ts'])
df
```

***1. Process the "start_time_ts" column to create "time_spec_points" column. Allot exact points equaling the time of the day the user is surfing, but allot extra 5 points if they are surfing after 4 o clock in the afternoon. For example: If the user starts surfing at 9 o clock in the morning allot 9 points, but user surfing after 4 o clock in the afternoon will receive 21 points.***

```python
df['hour']=df['start_time_ts'].dt.hour
df.head(2)
```

```python
def Time_spec(t):
    if (t<16):
        return t
    elif t>=16:
        return t+5
```

```python
df['time_spec_points']=df['hour'].apply(lambda x:Time_spec(x))
df.head()
```

***2. Reduce the "total_duration" round to 2 decimal places: example (166.711175 -> 166.71) (0.935320 -> 0.94)***

```python
df['total_duration']=df['total_duration'].apply(lambda x:float(np.round(x, 2)))
df.head()
```

***3.  Add the values of columns  "total_clicks", "total_items", "total_cats" to create a new column: "total_inventory". If the value "total_inventory" is greater than 10, then make sure the value is increased by 100%.***

```python
def get_total_inventory(clicks,items,cats):
    total=clicks+items+cats
    if total>10:
        return total*2
    else:
        return total
```

```python
df['total_inventory'] = df[['total_clicks','total_items','total_cats']].apply(lambda x: get_total_inventory(*x), axis=1)
df.head()
```

***4. Create a new column "give_big_discount". The value of '"give_big_discount" should be 1 if day_of_week is 0 and is_special_day is 1, otherwise 0.***

```python
def get_big_discount(DayOfWeek,Special_day):
    if (DayOfWeek==0) & (Special_day==1):
        return 1
    else:
        return 0
    
df['give_big_discount'] = df[['day_of_week','is_special_day']].apply(lambda x: get_big_discount(*x), axis=1)
df.head()
```

***5. Create a new column prod_views_buys_ratio which should be a ratio of prod_views_freqs and prod_buys_freqs***

```python
df['prod_views_buys_ratio'] = df.prod_views_freqs.div(df.prod_buys_freqs)
df.head()
```

***6. Create a new column create loyalty_points which should be calculated based on the following conditions: if a user is spending more than 3 seconds on a Sunday -- and loyalty points score will be 10 multiplied seconds over 3.***

```python
df['day_name']=df['start_time_ts'].dt.day_name()
df.head()
```

```python
def get_loyalty_points(day_name,total_duration):
    if (day_name=='Sunday') & (total_duration>3):
        return 10*(int(total_duration) - 3) 
    else:
        return 0
    
df['loyalty_points'] = df[['day_name','total_duration']].apply(lambda x: get_loyalty_points(*x), axis=1)
df.head()
```
